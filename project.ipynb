```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
df = pd.read_csv('super.csv')
df
```

```python
df.info()
```

Rename the columns

```python
df.rename(columns={'Invoice ID':'Invoice_ID','Customer type':'Customer_type','Product line':'Product_line','Unit price':'Unit_price','Quantity':'Quantity','Tax 5%':'Tax'},inplace=True)
df.info()
```

clean the format of the column('Unit_price') to convert to float

```python
df['Unit_price']=df['Unit_price'].str.replace(' USD','')
df['Unit_price']=df['Unit_price'].astype(float)
df.info()
```

calculate the tax column to fill the null values

```python
df['Tax']=df['Unit_price']*df['Quantity']*0.05
df.head()
```

Calculate the Total column to clear the null values

```python
df['Total']=df['Unit_price']*df['Quantity']+df['Tax']
df.head()
```

```python
df.info()
```

clean the time column

```python
df['Time']= df['Time'].str.replace(' - ',':')
df
```

clean the time column


```python
df['Time']=pd.to_datetime(df['Time']).dt.time
df.head()
```

clean the date column

```python
df['Date']=pd.to_datetime(df['Date'])
df.head()
```

combine the date and time in one column with datetime format

```python
df['Date_time'] = pd.to_datetime(df['Date']) + pd.to_timedelta(df['Time'].astype(str))
df = df.drop(['Date', 'Time'], axis=1)
df.head()

```

```python
df.info()
```

create a city column that combine the columns 'Yangon','Naypyitaw','Mandalay'

```python
def city(row):
  if row['Yangon'] == 1:
    return 'Yangon'
  elif row['Naypyitaw'] == 1:
    return 'Naypyitaw'
  elif row['Mandalay'] == 1:
    return 'Mandalay'
  else:
    return 'Unknown'

df['City'] = df.apply(city, axis=1)
df = df.drop(['Yangon','Naypyitaw','Mandalay'], axis=1)
df.head()

```

the data is cleaned and ready to be visualised
